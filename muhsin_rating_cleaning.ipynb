```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r"C:\Users\pk\Downloads\merged.csv")
```

```python
#agency
def extract_credit_agency(row):
    row = str(row)
    agency = ''
    if 'CRISIL' in row:
        agency = 'CRISIL'
    elif 'ICRA' in row:
        agency = 'ICRA'
    elif 'CARE' in row:
        agency = 'CARE'
    elif 'ACUITE' in row:
        agency = 'ACUITE'
    elif 'IND' in row:
        agency = 'IND'
    elif 'BWR' in row:
        agency = 'BWR'
    elif 'Infomerics Valuation and Rating Pvt. Ltd' in row:
        agency = 'Infomerics Valuation and Rating Pvt. Ltd'
    if agency == '':
        agency = 'other'
    
    return agency
```

```python
# Define a function to extract the desired credit rating
def extract_credit_rating(row):
    row = str(row)
    rating = ''
    if 'CRISIL' in row:
        rating = row.split('CRISIL')[-1].split('/')[0].split(';')[0].strip()
    elif 'ICRA' in row:
        rating = row.split('ICRA')[-1].split('/')[0].split(';')[0].strip()
    elif 'CARE' in row:
        rating = row.split('CARE')[-1].split('/')[0].split(';')[0].strip()
    elif 'ACUITE' in row:
        rating = row.split('ACUITE')[-1].split('/')[0].split(';')[0].strip()
    elif 'IND' in row:
        rating = row.split('IND')[-1].split('/')[0].split(';')[0].strip()
    elif 'BWR' in row:
        rating = row.split('BWR')[-1].split('/')[0].split(';')[0].strip()
    elif 'Infomerics Valuation and Rating Pvt. Ltd' in row:
        rating = row.split('Infomerics Valuation and Rating Pvt. Ltd')[-1].split('/')[0].split(';')[0].strip()
    if rating == '':
        rating = row
        
    rating = re.sub(r'\(.*?\)|\[.*?\]', '', rating)
    rating = rating.replace('STABLE', '')
    rating = rating.replace(' Negative', '-')
    rating = rating.replace('Stable', '')
    rating = rating.replace(r'/Stable', '')
    rating = rating.replace(r'r', '')
    rating = rating.replace(r'/', '')
    rating = rating.replace('IRR', '')
    rating = rating.replace('IVR', '')
    rating = rating.replace('PP-MLD', '')
    rating = rating.strip()
    return rating
```

```python
df['company_credit_rating'] = df['creditrating'].apply(extract_credit_rating)
```

```python
df['agency'] = df['creditrating'].apply(extract_credit_agency)
```
