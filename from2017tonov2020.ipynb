```python
#Loading required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import numpy as np

from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
from datetime import datetime, timedelta

start_date = datetime(2017, 10, 12)
end_date = datetime(2020, 11, 16)

date_format = "%d-%b-%Y"

date_list = []
current_date = start_date

while current_date <= end_date:
    if current_date.weekday() not in [5, 6]:  # Exclude Saturday (5) and Sunday (6)
        formatted_date = current_date.strftime(date_format)
        date_list.append(formatted_date)
    current_date += timedelta(days=1)
```

```python
url = "https://www.fpi.nsdl.co.in/web/Reports/traderepositoryreport.aspx"
```

```python
#setting up selenium chrome driver
from selenium.webdriver.chrome.service import Service
s = Service("C:\\chromedriver.exe")
driver = webdriver.Chrome(service = s)
driver.get(url)
```

```python
#for i in date_list:
for i in date_list:
    e = driver.find_element(By.ID, "hdndrpview")
    driver.execute_script("arguments[0].setAttribute('value','1')", e)
    e = driver.find_element(By.ID, "hdnFromDate")
    driver.execute_script(f"arguments[0].setAttribute('value', '{i}')", e)
    driver.find_element(By.ID, "btnSubmit").click()
    
    time.sleep(3)
    
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    div=soup.select_one("div#dvTradeData")
    table=pd.read_html(str(div))
    
    if table[0].empty == False:
        table1 = table[0]
        table1['Date'] = i
        nov17 = nov17.append(table1.droplevel(0,axis = 1))

    else:
        continue
```
